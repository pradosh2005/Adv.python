```python
##class creation 
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
class wikipedia:
    def __init__(self, url):
        self.url = url 
        self.headers = {  
                  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/118.0.0.0 Safari/537.36"
}
        
    def get_page(self):
        res = requests.get(self.url, headers= self.headers)
        res.raise_for_status()
        return res.content


##craete method to soup 
    def get_soup(self):
        soup = BeautifulSoup(self.get_page())
        return soup 


## create method for title 
    def get_title(self):
        soup = self.get_soup()
        title = soup.find_all(name = 'title')
        return title.text

    def get_h1_tags(self):
        soup = self.get_soup()
        h1_tags = soup.find_all(name = 'h1')
        return h1_tags[0].text

    def get_h2_tags(self):
        soup = self.get_soup()
        h2_tags = soup.find_all(name = 'h2')
        return [i.text for i in h2_tags]

    def get_h3_tags(self):
        soup = self.get_soup()
        h3_tags = soup.find_all(name ='h3')
        return [i.text for i in h3_tags]

    def get_paras(self):
        soup = self.get_soup()
        paras = soup.find_all(name = 'p')
        for i in paras:
            print('------------')
            print(i.text)
            print('---------')

    def get_tables(self):
        soup  = self.get_soup()
        table = soup.find_all(name = 'table' , class_ = 'wikipedia')
        for i in table:
            print(pd.read_html(str(i)))
     
    def get_image_link(self):
        soup = self.get_soup()
        image = soup.find_all(name = 'a' , class_ = 'mw-file_description')
        return [image[i].get('herf') for i in  range(len(image))]


```
